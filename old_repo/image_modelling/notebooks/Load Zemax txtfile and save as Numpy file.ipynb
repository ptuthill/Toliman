```python
# Derived from snippet from Barnaby
import numpy as np
import codecs
import os

# Source file location
datapath = '../zemax_examples/'
filename = 'toliman-RC-f50_ParaxialVersion_psf_0deg_600nm.txt'

with codecs.open(datapath+filename, encoding='utf-16') as f:
    data = np.loadtxt(f, skiprows=21)
```

```python
# Check it looks good
import matplotlib.pyplot as plt
lolim = -5
sz=256
cnt = 512
subim = data[cnt-sz:cnt+sz, cnt-sz:cnt+sz]
plt.imshow(np.log10(subim), interpolation='Nearest', clim=[lolim,0], cmap='viridis')
plt.colorbar()
plt.show()
```

```python
# Export it as a more convenient (and compact) numpy binary file
np.save(datapath+os.path.splitext(filename)[0], data)
```
