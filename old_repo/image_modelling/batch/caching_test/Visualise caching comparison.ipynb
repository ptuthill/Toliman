```python
import json
with open('caching_comparison_results.txt', 'r') as file:
    results = json.load(file)
```

# Graphing improvements using caching

Remember to run `batch.py` first

```python
import numpy as np
import matplotlib.pyplot as plt

N = 2
ind = np.arange(N)  # the x locations for the groups
width = 0.27       # the width of the bars

fig = plt.figure()
ax = fig.add_subplot(111)

runs = ['mono', 'poly']
uncached = ax.bar(ind, 
              [results[s+'_uncached'] for s in runs], 
              width, color='r')
savecached = ax.bar(ind+width, 
              [results[s+'_savecached'] for s in runs], 
              width, color='g')
usecached = ax.bar(ind+2*width, 
              [results[s+'_usecached'] for s in runs], 
              width, color='b')

ax.set_ylabel('Time (s)')
ax.set_xticks(ind+width)
ax.set_xticklabels( runs )
ax.legend( (uncached[0], savecached[0], usecached[0]), ('Caching off', 'Saving cache', 'Using cache'),loc='upper left' )
ax.set_xlim(0-0.1,N+3*width+0.1-1)
plt.show()
```
