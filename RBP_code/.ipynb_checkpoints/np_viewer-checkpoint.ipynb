```python
ext = "/data/laser_650/test_{}_{}.npy"
```

```python
root = os.getcwd()
```

```python
im = np.load(root + ext.format(10, 1))
```

```python
c1 = im.shape[1]//2 # x centre
c2 = im.shape[2]//2 # y centre
s1 = 200 # Central size
s2 = 200 # Corner size
off = (c1-c2)//2 #offset
order = ["Red", "Green 1", "Green 2", "Blue"]
```

```python
fig = plt.figure(figsize=(15,10))

for i in range(4):
    plt.subplot(2, 2, i+1)
    plt.title("{}".format(order[i]))
    plt.imshow(im[i][c1-s1:c1+s1, c2-s1:c2+s1])
#     plt.imshow(rgb[i])
    plt.colorbar()
    
plt.show()
```

```python
fig = plt.figure(figsize=(7,5))

for i in range(4):
    plt.subplot(2, 2, i+1)
    plt.title("{}".format(order[i]))
    plt.imshow(im[i][0:s2, off:s2+off])
    plt.colorbar()
    
plt.show()
```
